## RailCard:

```python
df["Railcard"] = df["Railcard"].fillna("No Card")
df["Railcard"].value_counts()
```

## Reason of Delay:

```python
df["Reason for Delay"].value_counts()
```

```python
df.loc[df["Reason for Delay"] == "Weather"  ,"Reason for Delay" ] = "Weather Conditions"
df.loc[df["Reason for Delay"] == "Signal failure"  ,"Reason for Delay" ] = "Signal Failure"
df.loc[df["Reason for Delay"] == "Staffing"  ,"Reason for Delay" ] = "Staff Shortage"

df["Reason for Delay"]= df["Reason for Delay"].fillna('No Delay')
```

```python
df["Reason for Delay"].value_counts()
```

```python
df["Reason for Delay"].isna().sum()
```

## Reservation peroid:

```python
df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'])
df['Date of Journey'] = pd.to_datetime(df['Date of Journey'])

df['Reservation Period'] = (df['Date of Journey'] - df['Date of Purchase']).dt.days

print(df[['Date of Purchase', 'Date of Journey', 'Reservation Period']].head())
```

## Calculation of Delay:

```python
df_cleaned = df
df_cleaned["Actual Arrival Time"] = pd.to_timedelta(df["Actual Arrival Time"])
df_cleaned["Arrival Time"] = pd.to_timedelta(df["Arrival Time"])

df["Delay In Minutes"] = (df_cleaned["Actual Arrival Time"] - df_cleaned["Arrival Time"]).dt.total_seconds() / 60

df["Delay In Minutes"] = df["Delay In Minutes"].apply(lambda x: x + 1440 if x < 0 else x)

df["Delay In Minutes"].head()


```

## Day of Journey:

```python
df['Date of Journey'] = pd.to_datetime(df['Date of Journey'])

df['Day of Journey'] = df['Date of Journey'].dt.day_name()

df['Day of Journey'].value_counts()
```
